```python
import geopandas as gpd
import pandas as pd
import mplleaflet
import matplotlib.pyplot as plt
from shapely.geometry import Point, LineString

file = "data/Takeout/Fit/Activities/2019-06-09T15_23_55Z_PT1H24M25S_Mountain biking.csv"

def to_geo(df, x_field='longitude', y_field='latitude'):
    geometry = [Point(xy) for xy in zip(df[x_field], df[y_field])]
    return gpd.GeoDataFrame(df, geometry=geometry)

def to_line_geo(df, x_field='longitude', y_field='latitude'):
    points = [Point(xy) for xy in zip(df[x_field], df[y_field])]
    return gpd.GeoDataFrame(pd.DataFrame([LineString(points)], columns=['geometry']))

df = pd.read_csv(file).dropna()

df.head()
```

```python
line_df = to_line_geo(df)

fig,ax=plt.subplots(figsize=(10,8))

line_df.plot(ax=ax) #, markersize=10, marker='o', column='trip_count', cmap='coolwarm')

points_df = to_geo(df.iloc[::20, :])

points_df.plot(ax=ax, markersize=50, marker='o', column='distance', cmap='coolwarm')

mplleaflet.display(fig=fig, crs=line_df.crs, tiles='cartodb_positron')
```
